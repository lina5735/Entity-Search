```python
import spacy
from textblob import TextBlob
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize  
import os
import re
check=check1+check2
name_pattern1 = r'[A-Z][a-z]*(?: [A-Z][a-z]*)?(?: [A-Z][a-z]*)?'
name_pattern2 = r'[A-Z][a-z]*(?: [A-Z][a-z]+[A-Z][a-z]*)*(?: [A-Z][a-z]+[A-Z][a-z]*)*'
nlp = spacy.load("en_core_web_sm")
results = {'ID': [], 'People': [], 'Time': [], 'Organization': [], 'Location': []}
filtered_time = {'week', 'this', 'year','years','monday','tuesday','wednesday','friday','thursday','saturday','sunday',
                 "january", "february", "march", "april", "may", "june", "july", "august", "september",
                 "october", "november", "december",'that'}
for i, document_dict in enumerate(check):
    document = document_dict.get('Content', '')
    doc = nlp(document)
    blob = TextBlob(document)
    document2 = " ".join(blob.words)
    combined_names = set(re.findall(name_pattern1,document2) + re.findall(name_pattern2,document2))
    people = []
    location = []
    organization = []
    time = []
    for word in combined_names:
        for ent in doc.ents:
            if word==ent.text and ent.label_ == "PERSON" and word not in people:
                people.append(word)
    for word in blob.words:
        for ent in doc.ents:
            if word==ent.text and ent.label_ == "DATE" and word not in time:
                if ent.text.lower() not in filtered_time:
                    time.append(ent.text)
            elif word==ent.text and ent.label_ == "ORG" and word not in organization:
                organization.append(ent.text)
            elif word==ent.text and ent.label_ == "GPE" and word not in location:
                location.append(ent.text)
    results['ID'].append(document_dict.get('ID', ''))
    results['People'].append(people)
    results['Time'].append(time)
    results['Organization'].append(organization) 
    results['Location'].append(location)            
df = pd.DataFrame.from_dict(results, orient='columns')
df
```

```python
def convert_to_set(name_string):
    if pd.isna(name_string):
        return set()
    name_list = name_string.split(',')
    cleaned_list = [word.strip() for word in name_list if word.strip() != '']
    return set(cleaned_list)

for columns in {'People','Location','Organization'}:
    combined[columns] = combined[columns].apply(convert_to_set)
import datetime

def convert_time_to_set(time_item):
    if pd.isna(time_item):
        return set()
    elif isinstance(time_item, datetime.datetime):
        return {time_item.strftime('%Y-%m-%d')}
    elif isinstance(time_item, str):
        time_list = time_item.split(',')
        time_list = [time.strip() for time in time_list if time.strip()!='']
        return set(time_list)
    else:
        return set()
combined['Time'] = combined['Time'].apply(convert_time_to_set)

combined
```

```python
df = df.fillna('')
df['People'] = df['People'].fillna('')
df['Location'] = df['Location'].fillna('')
df['Organization'] = df['Organization'].fillna('')
df['Time'] = df['Time'].fillna('')
combined = combined.fillna('')
combined['People'] = combined['People'].fillna('')
combined['Location'] = combined['Location'].fillna('')
combined['Organization'] = combined['Organization'].fillna('')
combined['Time'] = combined['Time'].fillna('')
```

```python
results = []

for i in range(0, 250):
    result = {'ID': df['ID'][i]}

    for column in ['People', 'Organization', 'Location','Time']:
        system_labels = set(df[column][i])
        manual_labels = set(combined[column][i])
        TP = len(system_labels & manual_labels)
        FN = len(manual_labels - system_labels)
        FP = len(system_labels - manual_labels)
        precision = TP / (TP + FP) if TP + FP > 0 else 0
        recall = TP / (TP + FN) if TP + FN > 0 else 0
        f1 = 2 * precision * recall / (precision + recall) if precision + recall > 0 else 0
        result[f'{column}_Precision'] = precision
        result[f'{column}_Recall'] = recall
        result[f'{column}_F1'] = f1
    results.append(result)
results_table = pd.DataFrame(results)
results_table
```
